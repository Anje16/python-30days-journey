```python
# 2. DATA TYPES - Python figures these out automatically

# String (text)
message = "Hello World"
print(type(message))  # <class 'str'>

# Integer (whole number)
score = 100
print(type(score))    # <class 'int'>

# Float (decimal)
price = 19.99
print(type(price))    # <class 'float'>

# Boolean (True/False)
is_active = True
print(type(is_active))  # <class 'bool'>
```

```python
# 3. MATH OPERATIONS

x = 10
y = 3

print("Addition:", x + y)        # 13
print("Subtraction:", x - y)     # 7
print("Multiplication:", x * y)  # 30
print("Division:", x / y)        # 3.333...
print("Integer Division:", x // y)  # 3 (rounds down)
print("Remainder:", x % y)       # 1
print("Power:", x ** y)          # 1000 (10^3)
```

```python
# 4. F-STRINGS - Modern way to format text

name = "Jammie"
age = 28
city = "Berlin"

# Old way (don't use this)
print("My name is " + name + " and I am " + str(age))

# NEW WAY - F-strings (use this!)
print(f"My name is {name} and I am {age}")
print(f"I live in {city}")

# Math inside f-strings
print(f"Next year I'll be {age + 1}")
print(f"In 10 years I'll be {age + 10}")
```

```python
# Simple Calculator

print("=== SIMPLE CALCULATOR ===\n")

# Get two numbers from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform all operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2

# Display results
print("\n--- RESULTS ---")
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} √ó {num2} = {multiplication}")
print(f"{num1} √∑ {num2} = {division}")
```

```python
# Age in different units

current_year = 2025
birth_year = int(input("What year were you born? "))

age_years = current_year - birth_year
age_months = age_years * 12
age_days = age_years * 365
age_hours = age_days * 24

print(f"\nYou are approximately:")
print(f"  {age_years} years old")
print(f"  {age_months} months old")
print(f"  {age_days} days old")
print(f"  {age_hours} hours old!")
```

```python
# Celsius to Fahrenheit converter

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32

print(f"\n{celsius}¬∞C = {fahrenheit}¬∞F")

# Bonus: Add a weather description
if fahrenheit > 86:
    print("It's HOT! üî•")
elif fahrenheit > 68:
    print("Nice weather! ‚òÄÔ∏è")
else:
    print("It's cold! ü•∂")
```

```python
# PERSONAL PROFILE GENERATOR

print("="*50)
print("      PERSONAL PROFILE GENERATOR")
print("="*50)

# Collect information
print("\nPlease answer these questions:\n")
name = input("1. Full name: ")
age = int(input("2. Age: "))
city = input("3. City: ")
hobby = input("4. Favorite hobby: ")
goal = input("5. Your 2025 goal: ")

# Calculate some fun facts
age_in_months = age * 12
age_in_2030 = age + 6

# Generate profile
print("\n" + "="*50)
print("           YOUR PROFILE")
print("="*50)
print(f"\nName: {name}")
print(f"Age: {age} years ({age_in_months} months)")
print(f"Location: {city}")
print(f"Hobby: {hobby}")
print(f"2024 Goal: {goal}")
print(f"\nFun Fact: In 2030, you'll be {age_in_2030} years old!")
print("\n" + "="*50)
print("Profile generated successfully! ‚úÖ")
```

```python
# Python's dynamic typing in action:

x = 10          # x is int
print(type(x))  # <class 'int'>

x = "hello"     # Now x is string (same variable!)
print(type(x))  # <class 'str'>

x = 3.14        # Now x is float
print(type(x))  # <class 'float'>

# This flexibility is powerful but requires careful coding!
```
